# Python List Comprehension

Denne notebook giver en kort introduktion til **list comprehensions i Python**.
Formålet er at vise, hvordan man kan skrive kortere og mere læsbar kode
sammenlignet med traditionelle `for`-loops.


## Hvad er en list comprehension?

En list comprehension er en **kort måde at oprette lister på** i Python.
Den kan ofte erstatte et almindeligt `for`-loop.

```python
# Traditionelt for-loop
numbers = []
for x in range(5):
    numbers.append(x)

numbers
```

Samme eksempel skrevet med **list comprehension**:


```python
numbers = [x for x in range(5)]
numbers
```

## Syntaks

Grundformen for en list comprehension ser sådan ud:

[udtryk for element in iterabel if betingelse]

## Eksempel: Beregning på hvert element

Vi ganger hvert tal med 2.


```python
doubled = [x * 2 for x in range(5)]
doubled
```

## List comprehension med betingelse

Man kan også tilføje en `if`-betingelse.


```python
even_numbers = [x for x in range(10) if x % 2 == 0]
even_numbers
```

## Sammenligning: for-loop vs list comprehension


```python
# For-loop
evens = []
for x in range(10):
    if x % 2 == 0:
        evens.append(x)

# List comprehension
evens_comp = [x for x in range(10) if x % 2 == 0]

evens, evens_comp
```

## Hvornår skal man bruge list comprehensions?

✅ Når koden bliver kortere og mere læsbar  
❌ Ikke hvis logikken bliver for kompleks


## Konklusion

- List comprehensions gør koden kortere
- De er nemme at læse ved simple opgaver
- De bruges meget i Python og er gode at kende
